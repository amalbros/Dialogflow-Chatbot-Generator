```python
import pandas as pd
import numpy as np
d = pd.read_csv('Book1.csv',encoding='iso-8859-1' )
```

```python
df=d[1:]
df.head(30)
```

```python
g=df.groupby('Intent')
cn=g['Intent'].count()
print(cn)
```

```python
d=dict() 
d1=dict()
for intent,data in g:
    d[intent]=np.asarray(data[ 'User Says']).tolist()
    f=data['Text Response']
    d1[intent]=np.asarray(data['Text Response'].dropna(axis=0)).tolist()
for i in d1.keys():
    tt=d1[i]
    f=[]
    print(tt)
    for j in tt:
        j=j.replace('\'',"(u0027")
        f.append(j)
        print(f)
    d1[i]=f.copy()
for i in d.keys():
    print(i)
    print(d[i])
    print(d1[i])
```
